```python
# Dependencies
import pandas as pd
import os
```

```python
# Load in File from resources
# 'movie_scores.csv'

filePath = "movie_scores.csv"
```

```python
# Read and display with pandas
moviesDF = pd.read_csv(filePath)
moviesDF.head()
```

```python
# List all the columns the table provides
moviesDF.columns
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB
reducedMoviesDF = moviesDF[["FILM", "IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
reducedMoviesDF.head()
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
movies7 = reducedMoviesDF.loc[(reducedMoviesDF["IMDB"] >= 7.0), :]
print(movies7)
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
lesserKnown = movies7.loc[(movies7["IMDB_user_vote_count"] <=20000), :]
print(lesserKnown)
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
lesserKnown.to_csv("Output_goodMovies.csv", encoding="UTF-8", index=False, header=True)
```
