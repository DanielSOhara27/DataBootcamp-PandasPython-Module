```python
# Dependencies
import pandas as pd
```

```python
# File to load
filePath = "pokemon.csv"
```

```python
# Read with Pandas
pokemon_df = pd.read_csv(filePath, encoding="ISO-8859-1")
pokemon_df.head()
```

```python
#Getting rid of NaN
pokemon_df = pokemon_df.dropna(how="any")

# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
pokemon_df_extract = pokemon_df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_df_extract.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
pokemon_avg_stats_df = (pokemon_df_extract.groupby(["Type 1"])).mean().sort_index()
pokemon_avg_stats_df
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.
pokemon_avg_stats_df["Total Power"] = pokemon_avg_stats_df.sum(axis=1)
pokemon_avg_stats_df.reset_index(inplace=True)
pokemon_avg_stats_df
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
pokemon_avg_stats_df = pokemon_avg_stats_df.sort_values(by=["Total Power"], ascending=False)
```
